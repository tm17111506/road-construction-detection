```python
def agg_category_count(category_dict, anns):
    found = False
    for ann in anns:
        ann_cat_name = nusc_trainval.get('sample_annotation', ann)['category_name']
        if ann_cat_name in category_dict.keys():
            category_dict[ann_cat_name] += 1
            found = True
    return found

def agg_filenames(data):
    rslt = []
    for k, v in data.items():
        if 'CAM' in k:
            sample_data = nusc_trainval.get('sample_data', v)
            rslt.append(sample_data['filename'])
    return rslt
```

```python
sensor = 'CAM_FRONT'
filenames = []
category_of_interest = {'human.pedestrian.construction_worker': 0, \
                        'movable_object.trafficcone': 0, \
                        'vehicle.construction': 0, \
                        'movable_object.barrier': 0}

for i, scene in enumerate(nusc_trainval.scene):
    token = scene['first_sample_token']
    while (len(token) > 0):
        sample = nusc_trainval.get('sample', token)
        contains_category = agg_category_count(category_of_interest, sample['anns'])
        if contains_category:
            filenames += agg_filenames(sample['data'])
        token = sample['next']
    
#     cam_front_data = nusc_trainval.get('sample_data', sample['data'][sensor])
#     file_name = cam_front_data['filename']
#     filenames.append(file_name)
#     next_scene_token = cam_front_data['next']

#     while (len(next_scene_token) > 0):
#         sample_data = nusc_test.get('sample_data', next_scene_token)
#         file_name = sample_data['filename']
#         # print(file_name)
#         filenames.append(file_name)
#         next_scene_token = sample_data['next']
#         # print(next_scene_token)
```

```python
import json
from itertools import groupby

data_path = '/usr0/tma1/datasets/nuscenes/v1.0-test'

def key_func(k):
    return k['filename']

with open(data_path + '/image_annotations.json', 'r') as fd:
    data = json.load(fd)
    grouped_data = {k: list(v) for k, v in groupby(data, key_func)}
    
with open(data_path + "/image_annotations_grouped.json", 'w+') as fw:
    json.dump(grouped_data, fw)
    
```

```python
# Create a mini evaluation set
import numpy as np
import json

mini_set = []
num_classes = 23
classes = [0 for i in range(num_classes)]
val_path = '/usr0/tma1/datasets/nuscenes/nuscene_val_2.json'
with open(val_path, 'r') as fd:
    data = json.load(fd)

for d in data:
    annos=d['annotations']
    for anno in annos:
        if classes[anno['category_id']] < 2:
            mini_set.append(d)
            classes[anno['category_id']] += 1
            break
            
    if np.all(classes):
        break

with open('/usr0/tma1/datasets/nuscenes/nuscene_val_mini_2.json', 'w+') as fd:
    json.dump(mini_set, fd)
print("Done!")
```

```python
# Create cones only dataset
import copy
import os
import json

print("Filter to only traffic cones")
dataroot = '/usr0/tma1/datasets/nuscenes'
splits = ['train', 'val']
for split in splits:
    fd = open(os.path.join(dataroot, 'nuscene_{}_2.json'.format(split)), 'r')
    nuscenes_train = json.load(fd)

    nuscenes_traffic_cones = []

    for d in nuscenes_train:
        annotations = []
        for anno in d['annotations']:
            if anno['category_id'] == category_dict['movable_object.trafficcone']:
                anno_copy = copy.deepcopy(anno)
                anno['category_id'] = 0
                annotations.append(anno)
        if len(annotations) > 0:
            dp = copy.deepcopy(d)
            dp['annotations'] = annotations
            nuscenes_traffic_cones.append(dp)
        # print(dp)
    fd = open(os.path.join(dataroot, 'nuscenes_trafficcone_{}.json'.format(split)), 'w+')
    json.dump(nuscenes_traffic_cones, fd)
    fd.close()
print("Done!")
```
